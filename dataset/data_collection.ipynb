```python
import mediacloud.api
import json, datetime
from newsfetch.news import newspaper

import warnings
warnings.filterwarnings('ignore')

mc = mediacloud.api.MediaCloud('MC_API_KEY')
```

```python
def get_stories(mc, media_ids, fetch_size = 500):

  stories = []

  last_processed_stories_id = 0
  while len(stories) < 100:
      for m_ids in media_ids:
          fetched_stories = mc.storyList('media_id:'+str(m_ids),
                                        solr_filter=mc.publish_date_query(datetime.date(2015,1,1), datetime.date(2015,8,1)),
                                        last_processed_stories_id=last_processed_stories_id, rows= fetch_size)
          stories.extend(fetched_stories)
          if len( fetched_stories) < fetch_size:
              break
          last_processed_stories_id = stories[-1]['processed_stories_id']


  return stories
```

```python
conservative_stories = get_stories(mc, media_ids = [24669,1092])
print(json.dumps(conservative_stories))
```

```python
neutral_stories = get_stories(mc, media_ids = [1150])
print(json.dumps(neutral_stories))
```

```python
import csv

with open('annotation_dataset.csv', 'a', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)
    for s in range(len(conservative_stories)):
        news = newspaper(conservative_stories[s]['url'])
        writer.writerow([news.article,1])
```

```python
import csv

with open('annotation_dataset.csv', 'a', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)
    for s in range(len(neutral_stories)):
        news = newspaper(neutral_stories[s]['url'])
        writer.writerow([news.article,0])
```
