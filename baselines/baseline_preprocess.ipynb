```python
from spacy.lang.en import English
import spacy

nlp = English()
nlp.add_pipe(nlp.create_pipe('sentencizer')) 
nlp2 = spacy.load("en_core_web_sm")
# doc = nlp(raw_text)
# sentences = [sent.string.strip() for sent in doc.sents]

def make_sent(para):
  doc = nlp(para)
  sents = [sent.string.strip() for sent in doc.sents]
  sents2=[]
  for each in sents:
    doc2 = nlp2(each) 
    sent = ""
    for tok in doc2:
      sent+=tok.text+" "
    sents2.append(sent[:-1])
  return sents2
```

```python
  data['news'] = data["news"].apply(make_sent)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vect_0 = CountVectorizer(ngram_range=(1,3)) # bigram
count_vect_0.fit(sent_0.tolist())
sent_0_ngram_dict = count_vect_0.vocabulary_

count_vect_1 = CountVectorizer(ngram_range=(1,3)) # bigram
count_vect_1.fit(sent_1.tolist())
sent_1_ngram_dict = count_vect_1.vocabulary_
```

```python
sent_0_ngram_df = pd.DataFrame({'ngram' : sent_0_ngram_dict.keys(), 'count0' : sent_0_ngram_dict.values() })
sent_1_ngram_df = pd.DataFrame({'ngram' : sent_1_ngram_dict.keys(), 'count1' : sent_1_ngram_dict.values() })
```

```python
ngram_df = pd.merge(sent_0_ngram_df,sent_1_ngram_df,on='ngram',how='outer')
ngram_df = ngram_df.fillna(0)
```

```python
from copy import deepcopy
ngram_df_count0 =  deepcopy(ngram_df['count0'])
ngram_df_count1 =  deepcopy(ngram_df['count1'])
ngram_df['count0'] = (ngram_df_count0 + 1)/(ngram_df_count1 + 1)
ngram_df['count1'] = (ngram_df_count1 + 1)/(ngram_df_count0 + 1)
```
